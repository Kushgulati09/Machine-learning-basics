```python
import numpy as np
import matplotlib.pyplot as plt

def function(n,m,Σ):                                                     
    e = np.random.normal(loc = 0.0, scale = Σ, size = (n,1))          
    β = np.random.rand(m+1,1)                                           
    x = np.random.rand(n,m+1)                                               
    x[:,0] = 1
    y = np.matmul(x,β) + e   
    
    return y,x,β
```

```python
mean=0
Σ=0.20
n=9
m=10
#setting some random values
A=[]
A=function(n,m,Σ)
print(A)
```

```python
Y=A[0]
X=A[1]
B=A[2]
k=1000
T=0.0005
lr=0.01
prev_cost =0

print ("previous β=",B)


def cost(LR,k,X,Y,B):
    prev_cost =0
    for i in range(k):
        err = (np.matmul(X,B)-Y)
        cost = np.sum(err** 2) / (2 * n)
        if abs(cost - prev_cost) < T:
            break
        prev_cost  = cost
        gradient = 2/len(Y)* np.matmul(X.T, err)
        B -= LR*gradient
    return B, cost
     
final = cost(lr,k,X,Y,B)
print("Beta final list=",final[0])
print("cost=",final[1]) 
```
